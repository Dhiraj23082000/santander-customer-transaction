```
import numpy as np
import pandas as pd

from hyperopt import hp, tpe
from hyperopt.fmin import fmin

from sklearn.model_selection import cross_val_score, StratifiedKFold
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import make_scorer

import xgboost as xgb

import lightgbm as lgbm
```

```
#hyper opt for lightgbm
def objective(params):
    params = {
        'num_leaves': int(params['num_leaves']),
        'colsample_bytree': '{:.3f}'.format(params['colsample_bytree']),
        'learning_rate': '{:.3f}'.format(params['learning_rate']),
        'n_estimators': int(params['n_estimators']),
        'min_child_samples':int(params['min_child_samples']),
        'class_weight': params['class_weight'],
    }
    
    clf = lgbm.LGBMClassifier(
        n_jobs=-1,
        metrics = 'auc',
       max_depth = -1,
    **params
    )
    
    score = cross_val_score(clf, X_train_res, y_train_res, cv=5).mean()
    print("Gini {:.3f} params {}".format(score, params))
    return 1-score

space = {
    'num_leaves': hp.quniform('num_leaves', 8, 128,1),
    'colsample_bytree': hp.uniform('colsample_bytree', 0.3, 1.0),
    'learning_rate': hp.loguniform('learning_rate', np.log(0.005), np.log(0.2)),
    'n_estimators': hp.quniform('n_estimators', 25, 500, 50),
    'class_weight': hp.choice('class_weight', [None, 'balanced']),
    'min_child_samples': hp.quniform('min_child_samples', 20, 500, 5),
    
}

best = fmin(fn=objective,
            space=space,
            algo=tpe.suggest,
            max_evals=10)
```

```
#hyper opt for random forest
def objective(params):
    params = {'max_features': int(params['max_features']), 'n_estimators': int(params['n_estimators']), 'max_depth': int(params['max_depth']),'min_samples_split': int(params['min_samples_split']),'min_samples_leaf': int(params['min_samples_leaf'])}
    clf = RandomForestClassifier(n_jobs=4, class_weight='balanced', **params)
    score = cross_val_score(clf, X_train_res, y_train_res, scoring='roc_auc', cv=5).mean()
    print("Gini {:.3f} params {}".format(score, params))
    return 1-score

space = {
    'n_estimators': hp.quniform('n_estimators', 25, 500, 40),
    'max_depth': hp.quniform('max_depth', 1, 10, 1),
    'max_features': hp.quniform('max_features', 1,5,1),
    'min_samples_leaf': hp.quniform('min_samples_leaf', 1,20,2),
    'min_samples_split': hp.quniform('min_samples_split', 1,15,2),

}

best = fmin(fn=objective,
            space=space,
            algo=tpe.suggest,
            max_evals=5)
```

```


from sklearn.metrics import confusion_matrix
from matplotlib import pyplot as plt

conf_mat = confusion_matrix(y_true=y_test, y_pred=pred)
print('Confusion matrix:\n', conf_mat)

labels = ['Class 0', 'Class 1']
fig = plt.figure()
ax = fig.add_subplot(111)
cax = ax.matshow(conf_mat, cmap=plt.cm.Blues)
fig.colorbar(cax)
ax.set_xticklabels([''] + labels)
ax.set_yticklabels([''] + labels)
plt.xlabel('Predicted')
plt.ylabel('Expected')
plt.show()
```
